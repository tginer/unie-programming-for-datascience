## 1. What Is Functional Programming?

Functional programming is a **declarative** paradigm where we describe
*what* to compute rather than *how* to compute it step by step.

Core ideas:
- **Pure functions** — same input always produces the same output, no side effects
- **Small, composable functions** — combine simple pieces into complex pipelines
- **No mutation** — transform data and return new results instead of changing existing data

Python provides three built-in functions for functional programming:

| Function | Purpose | Returns |
|---|---|---|
| `filter(func, iterable)` | Keep elements where `func` returns `True` | Iterator of selected elements |
| `map(func, iterable)` | Apply `func` to every element | Iterator of transformed elements |
| `reduce(func, iterable)` | Combine all elements into one value | Single accumulated value |

---

## 2. `filter()`

`filter(function, iterable)` returns an iterator containing only the
elements for which the function returns `True`.

### 2.1 Examples 

Filter by Category:

````python
def is_romance(ep):
    return ep["category"] == "Romance"


romance_episodes = list(filter(is_romance, episodes))

print(f"Romance episodes ({len(romance_episodes)}):")
for ep in romance_episodes:
    print(f"  S{ep['season']:02d}E{ep['episode']:02d} - {ep['title']}")
````

Filter by Season:

````python
def is_early_season(ep):
    """True if the episode is from seasons 1-3."""
    return ep["season"] <= 3


early_eps = list(filter(is_early_season, episodes))
print(f"Early-season episodes (S1-S3): {len(early_eps)} out of {len(episodes)}")
````

Filter by Keyword in Description:

````python
def mentions_wedding(ep):
    return "wedding" in ep["description"].lower()


wedding_eps = list(filter(mentions_wedding, episodes))

print(f"Episodes mentioning 'wedding' ({len(wedding_eps)}):")
for ep in wedding_eps:
    print(f"  S{ep['season']:02d}E{ep['episode']:02d} - {ep['title']} [{ep['category']}]")
````

---

## 3. `map()`

`map(function, iterable)` applies the function to **every** element and
returns an iterator of the results.

### 3.1 Examples 
Extract episode codes:

````python
def episode_code(ep):
    return f"S{ep['season']:02d}E{ep['episode']:02d}"


codes = list(map(episode_code, episodes))
print("All episode codes:")
print(codes)
````

Map to Summaries

````python
def to_summary(ep):
    code = f"S{ep['season']:02d}E{ep['episode']:02d}"
    return f"{code} - {ep['title']} [{ep['category']}]"


summaries = list(map(to_summary, episodes))
for s in summaries[:5]:
    print(s)
````

Map to Word Counts

````python
def description_word_count(ep):
    return len(ep["description"].split())


word_counts = list(map(description_word_count, episodes))
print("Word counts per episode:")
print(word_counts)
print(f"\nMin: {min(word_counts)}, Max: {max(word_counts)}")
````

---

## 4. `reduce()`

`reduce(function, iterable)` takes a function of **two arguments** and
applies it cumulatively to the elements, reducing the iterable to a
single value.

`reduce` lives in the `functools` module (not built-in like `filter`/`map`).

### 4.1 Examples 
Total Word Count:

````python
from functools import reduce


def add(a, b):
    return a + b


total_words = reduce(add, word_counts)
print(f"Total words across all descriptions: {total_words}")
print(f"Average words per episode: {total_words / len(episodes):.1f}")
````

Find the Longest Description

````python
def longer_description(ep_a, ep_b):
    """Return the episode with the longer description."""
    if len(ep_a["description"]) >= len(ep_b["description"]):
        return ep_a
    return ep_b


longest = reduce(longer_description, episodes)
print(f"Longest description:")
print(f"  S{longest['season']:02d}E{longest['episode']:02d} - {longest['title']}")
print(f"  {len(longest['description'])} characters, {len(longest['description'].split())} words")
````

Build a Category Frequency Dictionary

````python
def count_categories(acc, ep):
    """Accumulate category counts into a dictionary."""
    cat = ep["category"]
    acc[cat] = acc.get(cat, 0) + 1
    return acc


category_freq = reduce(count_categories, episodes, {})
print("Category frequencies (built with reduce):")
for cat, count in sorted(category_freq.items()):
    print(f"  {cat:<12} {count:>2}  {'#' * count}")
````

## 5. Lambda Functions

A **lambda** is an anonymous function defined in a single expression.
Lambdas are often used as quick, throwaway functions inside
`filter()`, `map()`, and `reduce()`.

```python
lambda arguments: expression
```

````python
family_eps = list(filter(lambda ep: ep["category"] == "Family", episodes))
print(f"Family episodes (filter + lambda): {len(family_eps)}")

titles = list(map(lambda ep: ep["title"], episodes))
print(f"\nAll titles (map + lambda): {titles[:5]} ...")

total = reduce(lambda a, b: a + b,
               map(lambda ep: len(ep["description"].split()), episodes))
print(f"\nTotal words (reduce + lambda): {total}")
````

## 6. Composing a Pipeline

The real power of functional programming comes from **chaining**
`filter`, `map`, and `reduce` together.

**Task:** Find the average word count of Romance episode descriptions.

````python
romance_only = filter(lambda ep: ep["category"] == "Romance", episodes)

romance_word_counts = list(map(lambda ep: len(ep["description"].split()), romance_only))

total_romance_words = reduce(lambda a, b: a + b, romance_word_counts)

avg = total_romance_words / len(romance_word_counts)

print(f"Romance episodes  : {len(romance_word_counts)}")
print(f"Total words       : {total_romance_words}")
print(f"Average word count: {avg:.1f}")
````

**Compare:** The same task using a traditional `for` loop:

```python
total = 0
count = 0
for ep in episodes:
    if ep["category"] == "Romance":
        total += len(ep["description"].split())
        count += 1
avg = total / count
```

Both approaches produce the same result. The functional style makes each
step (filter → transform → reduce) explicit and composable.

### Pipeline Example 2 — Titles of Career Episodes with long descriptions

````python
career_eps = filter(lambda ep: ep["category"] == "Career", episodes)

long_career = filter(lambda ep: len(ep["description"].split()) > 30, career_eps)

titles = list(map(lambda ep: f"S{ep['season']:02d}E{ep['episode']:02d} - {ep['title']}",
                  long_career))

print(f"Career episodes with 30+ word descriptions ({len(titles)}):")
for t in titles:
    print(f"  {t}")
````

## Summary

| Function | Purpose | Input | Output |
|---|---|---|---|
| `filter(func, iter)` | Select elements | `func` returns `True`/`False` | Iterator of matching elements |
| `map(func, iter)` | Transform elements | `func` returns new value | Iterator of transformed values |
| `reduce(func, iter)` | Accumulate into one | `func(acc, elem)` returns new acc | Single value |
| `lambda args: expr` | Inline anonymous function | Any arguments | Single expression result |

**Key takeaway:** Functional programming lets you build data transformation
**pipelines** by chaining small, reusable functions — the same pattern used
by pandas (`.filter().apply().agg()`) and scikit-learn pipelines.
