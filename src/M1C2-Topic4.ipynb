## 1. Object-Oriented Programming (OOP)

OOP is a programming paradigm that organises code around **objects** rather than
functions and logic alone. An object combines:

- **Attributes** — the data it holds (like variables attached to the object)
- **Methods** — the functions it can perform (like functions attached to the object)

A **class** is the blueprint; an **object** (or **instance**) is a concrete realisation
of that blueprint.

| Concept | Analogy | In our dataset |
|---|---|---|
| Class | Cookie cutter | `Episode` |
| Object | A cookie | A specific episode like *"The Pilot"* |
| Attribute | Shape, colour | `season`, `episode`, `title`, `description`, `category` |
| Method | What you can do with it | `summary()`, `is_from_season()`, `matches_keyword()` |

### 1.1 Defining a Class — The `Episode` Blueprint

### 1.5 Instance Methods

Methods are functions defined inside a class. They always receive `self` as the
first parameter and can access any attribute of the object.

Let's enhance our `Episode` class with useful methods.

````python
class Episode:
    """Represents a single FRIENDS episode."""

    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category

    def episode_code(self):
        """Return a human-readable code like S01E07."""
        return f"S{self.season:02d}E{self.episode:02d}"

    def is_from_season(self, season_number):
        """Check whether this episode belongs to a given season."""
        return self.season == season_number

    def matches_keyword(self, keyword):
        """Check whether the keyword appears in the description (case-insensitive)."""
        return keyword.lower() in self.description.lower()

    def summary(self):
        """Return a one-line summary of the episode."""
        return f"{self.episode_code()} - {self.title} [{self.category}]"
````

````python
pilot = Episode(
    season=1, episode=1,
    title="The Pilot",
    description=(
        "Rachel arrives at Central Perk in a wedding dress after running away "
        "from her wedding. Monica lets Rachel move in and the whole group "
        "welcomes her. The friends support each other through the change."
    ),
    category="Friendship",
)

print("Summary        :", pilot.summary())
print("Episode code   :", pilot.episode_code())
print("Is from S1?    :", pilot.is_from_season(1))
print("Is from S5?    :", pilot.is_from_season(5))
print("Has 'wedding'? :", pilot.matches_keyword("wedding"))
print("Has 'dinosaur'?:", pilot.matches_keyword("dinosaur"))
````

### 1.6 Special Methods: `__str__`, `__repr__`, and `__eq__`

Python uses **special methods** (also called *dunder methods* — short for
"double underscore") to define how objects behave with built-in operations:

| Method | When Python calls it | Example |
|---|---|---|
| `__str__` | `print(ep)` or `str(ep)` | Human-friendly output |
| `__repr__` | `repr(ep)` or inside a list `[ep]` | Developer-friendly output that can recreate the object |
| `__eq__` | `ep_a == ep_b` | Determines whether two objects are considered equal |

Without these methods, printing an object shows something unhelpful like
`<__main__.Episode object at 0x7f...>`, and `==` only checks whether two
variables point to the *exact same object in memory*, not whether they
represent the same episode. Let's fix all of that.

````python
class Episode:
    """Represents a single FRIENDS episode."""

    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category

    def episode_code(self):
        """Return a human-readable code like S01E07."""
        return f"S{self.season:02d}E{self.episode:02d}"

    def is_from_season(self, season_number):
        """Check whether this episode belongs to a given season."""
        return self.season == season_number

    def matches_keyword(self, keyword):
        """Check whether the keyword appears in the description (case-insensitive)."""
        return keyword.lower() in self.description.lower()

    def summary(self):
        """Return a one-line summary of the episode."""
        return f"{self.episode_code()} - {self.title} [{self.category}]"

    def __str__(self):
        return self.summary()

    def __repr__(self):
        return (f"Episode(season={self.season}, episode={self.episode}, "
                f"title='{self.title}', category='{self.category}')")

    def __eq__(self, other):
        """Two episodes are equal if they have the same season and episode number."""
        if not isinstance(other, Episode):
            return False
        return self.season == other.season and self.episode == other.episode
````

````python
pilot = Episode(
    season=1, episode=1,
    title="The Pilot",
    description=(
        "Rachel arrives at Central Perk in a wedding dress after running away "
        "from her wedding. Monica lets Rachel move in and the whole group "
        "welcomes her. The friends support each other through the change."
    ),
    category="Friendship",
)

print("print() uses __str__:")
print(pilot)

print("\n__repr__ (developer view):")
print(repr(pilot))

print("\nInside a list, Python uses __repr__:")
print([pilot])

print("\n__eq__ (equality comparison):")
same_pilot = Episode(
    season=1, episode=1,
    title="The Pilot",
    description="Different description text here.",
    category="Friendship",
)
blackout = Episode(
    season=1, episode=7,
    title="The One with the Blackout",
    description="A citywide blackout hits New York...",
    category="Romance",
)

print(f"pilot == same_pilot? {pilot == same_pilot}  (same S01E01)")
print(f"pilot == blackout?   {pilot == blackout}  (S01E01 vs S01E07)")
print(f"pilot == 'text'?     {pilot == 'text'}  (different type)")
print(f"pilot is same_pilot? {pilot is same_pilot}  (different objects in memory)")
````

### 1.7 Class Attributes vs Instance Attributes

- **Instance attributes** belong to a specific object (set via `self.x = ...` inside `__init__`).
- **Class attributes** are shared by *all* instances of the class.

A common use case: tracking how many episodes we have created.

````python
class Episode:
    """Represents a single FRIENDS episode."""

    VALID_CATEGORIES = {"Romance", "Career", "Family", "Friendship"}
    total_created = 0

    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category
        Episode.total_created += 1

    def episode_code(self):
        return f"S{self.season:02d}E{self.episode:02d}"

    def is_from_season(self, season_number):
        return self.season == season_number

    def matches_keyword(self, keyword):
        return keyword.lower() in self.description.lower()

    def summary(self):
        return f"{self.episode_code()} - {self.title} [{self.category}]"

    def __str__(self):
        return self.summary()

    def __repr__(self):
        return (f"Episode(season={self.season}, episode={self.episode}, "
                f"title='{self.title}', category='{self.category}')")
````

````python
print("Valid categories:", Episode.VALID_CATEGORIES)
print("Episodes created so far:", Episode.total_created)

ep1 = Episode(1, 1, "The Pilot",
              "Rachel arrives at Central Perk...", "Friendship")
ep2 = Episode(1, 7, "The One with the Blackout",
              "A citywide blackout hits New York...", "Romance")

print("After creating 2 episodes:", Episode.total_created)
print("Accessed from ep1:", ep1.total_created)
print("Accessed from ep2:", ep2.total_created)
print("Same value — class attribute is shared.")
````

## 2. Inheritance & Encapsulation

**Inheritance** lets us create a new class that *extends* an existing one, inheriting
all attributes and methods while adding or overriding specific behaviour.

**Encapsulation** is the practice of restricting direct access to some attributes,
exposing them only through controlled methods (getters/setters).

### 2.1 Inheritance — Specialised Episode Types

Suppose we want `RomanceEpisode` objects to automatically know the main couple
involved. We create a **subclass** that inherits from `Episode` and adds an
extra attribute.

````python
class RomanceEpisode(Episode):
    """An Episode subclass that tracks the main romantic couple."""

    def __init__(self, season, episode, title, description, couple):
        super().__init__(season, episode, title, description, category="Romance")
        self.couple = couple

    def summary(self):
        base = super().summary()
        return f"{base} ❤ {self.couple}"


ross_rachel = RomanceEpisode(
    season=2, episode=7,
    title="The One Where Ross Finds Out",
    description=(
        "Rachel leaves a drunken voicemail confessing her feelings to Ross. "
        "Ross listens to the message and confronts Rachel. They share a "
        "passionate kiss at Central Perk, beginning their love story."
    ),
    couple="Ross & Rachel",
)

print(ross_rachel.summary())
print("Is an Episode?       ", isinstance(ross_rachel, Episode))
print("Is a RomanceEpisode? ", isinstance(ross_rachel, RomanceEpisode))
print("Couple attribute     :", ross_rachel.couple)
print("Category (inherited) :", ross_rachel.category)
````

`super()` calls the parent class's `__init__`, so we don't have to repeat all the
attribute assignments. The subclass only adds what is new (`couple`) and overrides
what is different (`summary`).

Key points:
- `RomanceEpisode` **inherits** `episode_code()`, `is_from_season()`, `matches_keyword()`, `__str__`, and `__repr__` from `Episode`.
- `RomanceEpisode` **overrides** `summary()` to include the couple.
- `isinstance(ross_rachel, Episode)` returns `True` — a `RomanceEpisode` **is** an `Episode`.

### 2.2 Encapsulation — Protecting Internal State

By convention, attributes prefixed with an underscore (`_`) are considered
"private" — not to be accessed directly from outside. We expose them through
**properties** (controlled getters and setters).

This lets us add validation: for example, ensuring the `category` is always
one of the valid values.

#### How to change the way an attribute is handled: modify default getter/setter

In essence, when we access `category`  or set `category`  value:
```
ep.category          # Python calls the @property method (getter)
ep.category = "X"    # Python calls the @category.setter method (setter)
```

The `@` symbol is a **decorator** — a Python feature that modifies the
behaviour of the function defined right below it. We can change it to add an additional validator before setting the `category`  value. 

| Syntax | What it does |
|---|---|
| `@property` | Turns a method into a **getter** — you can read `ep.category` as if it were a plain attribute, but Python actually calls the method behind the scenes |
| `@category.setter` | Turns a method into a **setter** — you can write `ep.category = "Romance"` and Python calls this method, giving us a place to add validation |

Without `@property`, you would need explicit methods like `ep.get_category()`
and `ep.set_category("Romance")`. The decorator lets you keep the clean
dot-notation syntax (`ep.category`) while still running custom code
under the hood.



````python
class Episode:
    """Represents a single FRIENDS episode with validated category."""

    VALID_CATEGORIES = {"Romance", "Career", "Family", "Friendship"}

    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category  # uses the setter below

    @property
    def category(self):
        """Get the episode category."""
        return self._category

    @category.setter
    def category(self, value):
        """Set category only if it is a valid value."""
        if value not in self.VALID_CATEGORIES:
            raise ValueError(
                f"Invalid category '{value}'. "
                f"Must be one of {sorted(self.VALID_CATEGORIES)}"
            )
        self._category = value

    def episode_code(self):
        return f"S{self.season:02d}E{self.episode:02d}"

    def summary(self):
        return f"{self.episode_code()} - {self.title} [{self.category}]"

    def __str__(self):
        return self.summary()

    def __repr__(self):
        return (f"Episode(season={self.season}, episode={self.episode}, "
                f"title='{self.title}', category='{self.category}')")
````

````python
ep = Episode(1, 1, "The Pilot",
             "Rachel arrives at Central Perk...", "Friendship")
print("Category:", ep.category)

ep.category = "Romance"
print("Updated :", ep.category)

ep.category= "Comedy"

# Handle the error properly with try/except - explained in Section 
# try:
#     ep.category = "Comedy"
# except ValueError as e:
#     print(f"\nValidation error: {e}")
````

The setter rejected `"Comedy"` because it is not in `VALID_CATEGORIES`.
Without encapsulation, nothing would prevent us from assigning invalid values.
This is a common pattern in professional codebases: keep the data consistent
by validating at the point of assignment.

### 3. Summary OOP — Episode with Multiple Features

Let's combine several OOP concepts into a richer example.
We add a method that returns the word count of the description,
a method to compare two episodes, and we demonstrate how objects
interact with each other.

````python
class Episode:
    """Full-featured Episode class."""

    VALID_CATEGORIES = {"Romance", "Career", "Family", "Friendship"}

    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category

    @property
    def category(self):
        return self._category

    @category.setter
    def category(self, value):
        if value not in self.VALID_CATEGORIES:
            raise ValueError(
                f"Invalid category '{value}'. "
                f"Must be one of {sorted(self.VALID_CATEGORIES)}"
            )
        self._category = value

    def episode_code(self):
        return f"S{self.season:02d}E{self.episode:02d}"

    def word_count(self):
        """Return the number of words in the description."""
        return len(self.description.split())

    def same_category_as(self, other):
        """Check if this episode shares a category with another."""
        return self.category == other.category

    def is_from_season(self, season_number):
        return self.season == season_number

    def matches_keyword(self, keyword):
        return keyword.lower() in self.description.lower()

    def summary(self):
        return f"{self.episode_code()} - {self.title} [{self.category}]"

    def __str__(self):
        return self.summary()

    def __repr__(self):
        return (f"Episode(season={self.season}, episode={self.episode}, "
                f"title='{self.title}', category='{self.category}')")

    def __eq__(self, other):
        """Two episodes are equal if they have the same season and episode number."""
        if not isinstance(other, Episode):
            return False
        return self.season == other.season and self.episode == other.episode
````

## 3. Error Handling

Errors are inevitable in programming. Python provides a structured way to
**catch**, **handle**, and even **raise** errors so that your program can recover
gracefully instead of crashing.

Error handling is especially important in Data Science: files might be missing,
data might have unexpected formats, and network requests can fail.

### 3.1 `try` / `except` — Catching Errors

The `try` block contains code that might fail. If an error occurs, Python
jumps to the matching `except` block instead of crashing.

````python
print("--- Example 1: Accessing an invalid index ---")
titles = ["The Pilot", "The Blackout", "The Last One"]

try:
    print(titles[10])
except IndexError as e:
    print(f"Caught IndexError: {e}")

print("\n--- Example 2: Division by zero ---")
total_episodes = 25
seasons_with_data = 0

try:
    avg = total_episodes / seasons_with_data
except ZeroDivisionError as e:
    print(f"Caught ZeroDivisionError: {e}")
    avg = 0

print(f"Average (safe): {avg}")

print("\n--- Example 3: Wrong type ---")
try:
    result = "Season" + 1
except TypeError as e:
    print(f"Caught TypeError: {e}")
````

### 3.2 Common Built-in Exception Types

| Exception | When it happens | FRIENDS example |
|---|---|---|
| `IndexError` | Index out of range | Accessing `episodes[100]` |
| `KeyError` | Missing dictionary key | `ep_dict["rating"]` when "rating" doesn't exist |
| `TypeError` | Wrong type in operation | `"Season" + 1` instead of `"Season" + str(1)` |
| `ValueError` | Right type, wrong value | `int("abc")` |
| `AttributeError` | Accessing non-existent attribute | `ep.rating` when Episode has no `rating` |
| `ZeroDivisionError` | Dividing by zero | Average when there are 0 episodes |
| `FileNotFoundError` | File does not exist | Opening a non-existent CSV |

````python
ep = Episode(1, 1, "The Pilot",
             "Rachel arrives at Central Perk...", "Friendship")

print("--- KeyError ---")
ep_dict = {"title": "The Pilot", "season": 1}
try:
    print(ep_dict["rating"])
except KeyError as e:
    print(f"Caught KeyError: {e}")

print("\n--- AttributeError ---")
try:
    print(ep.rating)
except AttributeError as e:
    print(f"Caught AttributeError: {e}")

print("\n--- ValueError ---")
try:
    season_num = int("one")
except ValueError as e:
    print(f"Caught ValueError: {e}")
````

### 3.3 `try / except / else / finally`

- `else` runs only if **no** exception was raised.
- `finally` runs **always**, whether or not an exception occurred. It is typically
  used for cleanup (closing files, releasing resources).

````python
def safe_get_episode(episode_list, index):
    """Safely retrieve an episode by index."""
    try:
        ep = episode_list[index]
    except IndexError:
        print(f"  Index {index} is out of range (max {len(episode_list) - 1})")
        return None
    else:
        print(f"  Found: {ep.summary()}")
        return ep
    finally:
        print(f"  (lookup complete for index {index})")


episodes_small = [
    Episode(1, 1, "The Pilot", "Rachel arrives at Central Perk...", "Friendship"),
    Episode(1, 7, "The One with the Blackout", "A citywide blackout...", "Romance"),
]

print("Valid index:")
safe_get_episode(episodes_small, 0)

print("\nInvalid index:")
safe_get_episode(episodes_small, 99)
````

### 3.4 Raising Exceptions

Use `raise` to signal that something went wrong. This is how libraries like
pandas raise `ValueError` when you pass invalid arguments.

We already saw this in section 2.2 with our `category` setter — let's
explore `raise` in more depth.

````python
def create_episode_from_dict(data):
    """Create an Episode from a dictionary, with validation."""
    required_keys = {"season", "episode", "title", "description", "category"}
    missing = required_keys - set(data.keys())

    if missing:
        raise KeyError(f"Missing required fields: {missing}")

    if not isinstance(data["season"], int) or data["season"] < 1:
        raise ValueError(f"Season must be a positive integer, got {data['season']}")

    if not isinstance(data["episode"], int) or data["episode"] < 1:
        raise ValueError(f"Episode must be a positive integer, got {data['episode']}")

    return Episode(**data)


print("--- Valid data ---")
valid = {"season": 1, "episode": 1, "title": "The Pilot",
         "description": "Rachel arrives...", "category": "Friendship"}
ep = create_episode_from_dict(valid)
print(f"Created: {ep}")

print("\n--- Missing field ---")
try:
    bad_data = {"season": 1, "title": "Oops"}
    create_episode_from_dict(bad_data)
except KeyError as e:
    print(f"Caught: {e}")

print("\n--- Invalid season ---")
try:
    bad_season = {"season": -1, "episode": 1, "title": "Bad",
                  "description": "...", "category": "Romance"}
    create_episode_from_dict(bad_season)
except ValueError as e:
    print(f"Caught: {e}")
````

### 4 - OOP and Error Handling — A Robust `EpisodeCollection`

Let's bring everything together by building an `EpisodeCollection` class
that manages a list of episodes with proper error handling at every step.

````python
class EpisodeCollection:
    """A managed collection of Episode objects."""

    def __init__(self):
        self._episodes = []

    def add(self, episode):
        """Add an episode. Raises TypeError if not an Episode, ValueError if duplicate."""
        if not isinstance(episode, Episode):
            raise TypeError(
                f"Expected an Episode object, got {type(episode).__name__}"
            )
        for existing in self._episodes:
            if existing == episode:
                raise ValueError(
                    f"{episode.episode_code()} already exists in the collection."
                )
        self._episodes.append(episode)

    def find(self, season, episode_num):
        """Find an episode by season and episode number."""
        if season < 1 or season > 10:
            raise ValueError(
                f"Season {season} is invalid. FRIENDS has seasons 1-10."
            )
        for ep in self._episodes:
            if ep.season == season and ep.episode == episode_num:
                return ep
        raise LookupError(
            f"S{season:02d}E{episode_num:02d} not found."
        )

    def filter_by_category(self, category):
        """Return episodes matching the given category."""
        if category not in Episode.VALID_CATEGORIES:
            raise ValueError(
                f"Unknown category '{category}'. "
                f"Valid: {sorted(Episode.VALID_CATEGORIES)}"
            )
        return [ep for ep in self._episodes if ep.category == category]

    def __len__(self):
        return len(self._episodes)

    def __iter__(self):
        return iter(self._episodes)

    def __str__(self):
        return f"EpisodeCollection({len(self)} episodes)"
````

````python
collection = EpisodeCollection()

collection.add(Episode(1, 1, "The Pilot",
                        "Rachel arrives at Central Perk...", "Friendship"))
collection.add(Episode(1, 7, "The One with the Blackout",
                        "A citywide blackout hits New York...", "Romance"))
collection.add(Episode(2, 14, "The One with the Prom Video",
                        "The group watches an old prom video...", "Romance"))

print(collection)
print()

for ep in collection:
    print(f"  {ep}")
````

````python
print("--- Adding a non-Episode object ---")
try:
    collection.add("Not an episode")
except TypeError as e:
    print(f"TypeError: {e}")

print("\n--- Adding a duplicate ---")
try:
    collection.add(Episode(1, 1, "The Pilot", "Duplicate...", "Friendship"))
except ValueError as e:
    print(f"ValueError: {e}")

print("\n--- Finding an episode ---")
try:
    found = collection.find(1, 7)
    print(f"Found: {found}")
except LookupError as e:
    print(f"Not found: {e}")

print("\n--- Finding a non-existent episode ---")
try:
    collection.find(5, 99)
except LookupError as e:
    print(f"LookupError: {e}")

print("\n--- Invalid season ---")
try:
    collection.find(15, 1)
except ValueError as e:
    print(f"ValueError: {e}")

print("\n--- Filtering by invalid category ---")
try:
    collection.filter_by_category("Horror")
except ValueError as e:
    print(f"ValueError: {e}")

print("\n--- Filtering by valid category ---")
romance = collection.filter_by_category("Romance")
print(f"Romance episodes ({len(romance)}):")
for ep in romance:
    print(f"  {ep}")
````
