## 1. Reading and Writing Files

`open(filename, mode, encoding)` is the built-in function for file I/O.

| Mode | Meaning |
|---|---|
| `'r'` | Read (default) |
| `'w'` | Write (creates or **overwrites**) |
| `'a'` | Append |
| `'r+'` | Read and write |

The `with` statement ensures the file is **automatically closed** when the
block finishes, even if an error occurs.

### 1.1 Writing a Text File
The following code snippet opens a new file "friends_titles.txt and writes line by line all episodes titles. 

Check the file content after you execute the cell.

```python
with open("friends_titles.txt", "w", encoding="utf-8") as f:
    for ep in episodes:
        line = f"S{ep['season']:02d}E{ep['episode']:02d} - {ep['title']}\n"
        f.write(line)

print("File written.")
print("File closed?", f.closed)
```

### 1.2 Reading a Text File
The following code snippet opens the just created file "friends_titles.txt" and fills in the empty variable 'content' with the file content. 

```python
with open("friends_titles.txt", encoding="utf-8") as f:
    content = f.read()

print("Full file content:")
print(content[:300], "...")
```

Read the file line by line using the strip() function:

```python
with open("friends_titles.txt", encoding="utf-8") as f:
    print("Reading line by line:")
    for i, line in enumerate(f):
        print(f"  Line {i}: {line.strip()}")
        if i >= 4:
            print("  ...")
            break
```

### 1.3 Appending to a File
Opening an existing file and writing on it automatically appends the content.

```python
with open("friends_titles.txt", "a", encoding="utf-8") as f:
    f.write("--- End of dataset ---\n")

with open("friends_titles.txt", encoding="utf-8") as f:
    lines = f.readlines()

print(f"Total lines: {len(lines)}")
print(f"Last line  : {lines[-1].strip()}")
```

## 2. CSV Files

The `csv` module reads and writes **comma-separated values** — the most
common format for tabular data.

### 2.1 Writing CSV with `csv.writer` (list rows)
writerow() prints ine by line the content

```python
import csv

with open("friends_episodes.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile, delimiter=",")
    writer.writerow(["season", "episode", "title", "category"])
    for ep in episodes:
        writer.writerow([ep["season"], ep["episode"], ep["title"], ep["category"]])

print("CSV written with csv.writer")
```

### 2.2 Reading CSV with `csv.reader`
Reader reads row by row the CSV content

```python
with open("friends_episodes.csv", newline="", encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    header = next(reader)
    print(f"Header: {header}")
    for row in reader:
        season, episode, title, category = row
        print(f"  S{int(season):02d}E{int(episode):02d} - {title} [{category}]")
        if int(season) >= 2:
            print("  ...")
            break
```

### 2.3 Writing CSV with `csv.DictWriter` (dictionary rows)
Using DictWriter writes all dictionary keys' values row by row using the writerow()

```python
fields = ["season", "episode", "title", "description", "category"]

with open("friends_full.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fields)
    writer.writeheader()
    for ep in episodes:
        writer.writerow(ep)

print("CSV written with csv.DictWriter (all 5 fields)")
```

### 2.4 Reading CSV with `csv.DictReader`

```python
with open("friends_full.csv", newline="", encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    loaded_episodes = list(reader)

print(f"Loaded {len(loaded_episodes)} episodes from CSV")
print(f"Keys per row: {list(loaded_episodes[0].keys())}")

sample = loaded_episodes[0]
print(f"\nFirst episode:")
print(f"  Title   : {sample['title']}")
print(f"  Season  : {sample['season']}")
print(f"  Category: {sample['category']}")
print(f"  Desc    : {sample['description'][:60]}...")
```

**Important:** `csv.DictReader` returns all values as **strings**.
Notice that `sample['season']` is `'1'` (a string), not `1` (an integer).
You must convert manually with `int()` if you need numbers.

## 3. JSON Files

The `json` module serializes Python objects to a **human-readable text format**
that is widely used for data exchange across languages and systems.

| Function | What it does |
|---|---|
| `json.dumps(obj)` | Convert to JSON **string** |
| `json.dump(obj, f)` | Write JSON to a **file** |
| `json.loads(s)` | Parse a JSON **string** |
| `json.load(f)` | Read JSON from a **file** |

### 3.1 Writing JSON
json.dumps() saves the JSON content to a variable - not a file.

```python
import json

json_string = json.dumps(episodes[0], indent=2)
print("Single episode as JSON string:")
print(json_string)
```

Creating a new jsn file would then look like:

```python
with open("friends_episodes.json", "w", encoding="utf-8") as f:
    json.dump(episodes, f, indent=2, ensure_ascii=False)

print("Full dataset written to friends_episodes.json")
```

### 3.2 Reading JSON

```python
with open("friends_episodes.json", encoding="utf-8") as f:
    loaded = json.load(f)

print(f"Loaded {len(loaded)} episodes from JSON")
print(f"Type: {type(loaded)}")
print(f"First episode type: {type(loaded[0])}")

sample = loaded[0]
print(f"\nTitle   : {sample['title']}")
print(f"Season  : {sample['season']} (type: {type(sample['season']).__name__})")
print(f"Category: {sample['category']}")
```

Unlike CSV, JSON **preserves data types**: integers stay as integers,
strings stay as strings. This makes JSON a better choice when type
fidelity matters.

## 4. Pickle — Python Object Serialization

**Pickle** serializes Python objects to a **binary format**. Unlike JSON, pickle
can save *any* Python object — including custom classes and their instances.

| | `json` | `pickle` |
|---|---|---|
| **Format** | Text (UTF-8) | Binary |
| **Human-readable** | Yes | No |
| **Works outside Python** | Yes | No |
| **Secure** | Yes | No — only load trusted data |
| **Custom classes** | No (only built-in types) | Yes |

### 4.1 Pickling a List of Dictionaries

```python
import pickle

with open("friends_episodes.pkl", "wb") as f:
    pickle.dump(episodes, f)

print("Dataset pickled to friends_episodes.pkl")

with open("friends_episodes.pkl", "rb") as f:
    loaded_pkl = pickle.load(f)

print(f"Loaded {len(loaded_pkl)} episodes from pickle")
print(f"First title: {loaded_pkl[0]['title']}")
print(f"Data identical? {loaded_pkl == episodes}")
```

### 4.2 Pickling Custom Objects

Pickle's real power is saving **custom class instances** — something
JSON cannot do.

```python
class Episode:
    def __init__(self, season, episode, title, description, category):
        self.season = season
        self.episode = episode
        self.title = title
        self.description = description
        self.category = category

    def __str__(self):
        return f"S{self.season:02d}E{self.episode:02d} - {self.title} [{self.category}]"


episode_objects = []
for ep in episodes:
    obj = Episode(ep["season"], ep["episode"], ep["title"],
                  ep["description"], ep["category"])
    episode_objects.append(obj)

print(f"Created {len(episode_objects)} Episode objects")
print(f"Type: {type(episode_objects[0])}")
print(f"First: {episode_objects[0]}")
```

```python
with open("friends_objects.pkl", "wb") as f:
    pickle.dump(episode_objects, f)

print("Episode objects pickled.")

with open("friends_objects.pkl", "rb") as f:
    loaded_objects = pickle.load(f)

print(f"Loaded {len(loaded_objects)} Episode objects from pickle")
print(f"Type: {type(loaded_objects[0]).__name__}")
for obj in loaded_objects[:5]:
    print(f"  {obj}")
```

The loaded objects are fully functional `Episode` instances with all
their attributes and methods preserved. This would **not** be possible
with JSON.

## Summary

| Topic | Key concepts |
|---|---|
| **Text files** | `open()`, `read()`, `write()`, `with` statement, modes `'r'`/`'w'`/`'a'` |
| **CSV** | `csv.writer` / `csv.reader` for lists; `csv.DictWriter` / `csv.DictReader` for dicts |
| **JSON** | `json.dump()` / `json.load()` for files; `json.dumps()` / `json.loads()` for strings |
| **Pickle** | `pickle.dump()` / `pickle.load()` — binary, any Python object, not human-readable |

**When to use each format:**

- **Text files** — simple logs, quick exports
- **CSV** — tabular data, spreadsheet compatibility
- **JSON** — structured data, web APIs, cross-language exchange
- **Pickle** — persisting Python-specific objects (custom classes, trained ML models)
