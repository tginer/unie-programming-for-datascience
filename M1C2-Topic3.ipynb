## 1. Function Argument Options

Python functions support several ways to receive arguments beyond
simple positional parameters.

### 1.1 Default Values

Parameters can have **default values**. If the caller omits them,
the default is used automatically.

````python
def format_episode(ep, show_description=False, uppercase_title=False):
    """Format an episode dict with configurable options."""
    code = f"S{ep['season']:02d}E{ep['episode']:02d}"
    title = ep["title"].upper() if uppercase_title else ep["title"]
    result = f"{code} - {title} [{ep['category']}]"
    if show_description:
        result += f"\n   {ep['description'][:80]}..."
    return result


print("Defaults:")
print(format_episode(episodes[0]))

print("\nWith description:")
print(format_episode(episodes[0], show_description=True))

print("\nUppercase title:")
print(format_episode(episodes[0], uppercase_title=True))

print("\nBoth options:")
print(format_episode(episodes[0], show_description=True, uppercase_title=True))
````

### 1.2 `*args` — Variable Number of Positional Arguments

`*args` collects any number of positional arguments into a **tuple**.

````python
def count_keyword_hits(*keywords):
    """Count how many episodes mention ANY of the given keywords."""
    print(f"Received {len(keywords)} keywords (type: {type(keywords).__name__})")
    hits = 0
    for ep in episodes:
        desc = ep["description"].lower()
        if any(kw.lower() in desc for kw in keywords):
            hits += 1
    return hits


print("Hits for 'wedding':       ", count_keyword_hits("wedding"))
print("Hits for 'love','kiss':   ", count_keyword_hits("love", "kiss"))
print("Hits for 'Ross','Rachel': ", count_keyword_hits("Ross", "Rachel"))
````

### 1.3 `**kwargs` — Variable Number of Keyword Arguments

`**kwargs` collects any number of keyword arguments into a **dictionary**.

````python
def filter_episodes(data, **criteria):
    """Filter episodes by any combination of field values."""
    print(f"Filter criteria received: {criteria}")
    results = data
    for key, value in criteria.items():
        results = [ep for ep in results if ep.get(key) == value]
    return results


romance = filter_episodes(episodes, category="Romance")
print(f"Romance episodes: {len(romance)}")

s1_career = filter_episodes(episodes, season=1, category="Career")
print(f"Season 1 Career episodes: {len(s1_career)}")
for ep in s1_career:
    print(f"  S{ep['season']:02d}E{ep['episode']:02d} - {ep['title']}")
````

## 2. Functions Are Objects

In Python, functions are **first-class objects**. This means they can be:

- Assigned to variables
- Stored in lists or dictionaries
- Passed as arguments to other functions

### 2.1 Assigning Functions to Variables and Storing in a List

````python
def shout(text):
    return text.upper() + "!!!"

def whisper(text):
    return text.lower() + "..."

def title_case(text):
    return text.title()


my_func = shout
print(my_func(episodes[0]["title"]))

transformations = [shout, whisper, title_case]
sample_title = episodes[2]["title"]

print(f"\nOriginal: {sample_title}")
for func in transformations:
    print(f"  {func.__name__:>12}: {func(sample_title)}")
````

### 2.2 Strategy Pattern — Functions in a Dictionary

Storing functions in a dictionary lets us select behaviour at runtime
by name. This is known as the **strategy pattern**.

````python
STOP_WORDS = {
    "a", "an", "the", "and", "or", "but", "in", "on", "at", "to", "for",
    "of", "with", "by", "is", "are", "was", "were", "be", "been",
    "have", "has", "had", "do", "does", "did", "not", "no",
    "it", "its", "he", "she", "they", "them", "his", "her", "their",
    "this", "that", "who", "whom", "which", "what", "how", "all",
    "each", "every", "other", "some", "about", "up", "out", "so",
    "from", "into", "through", "after", "as", "over",
}


def normalize_basic(text):
    return text.lower().strip()


def normalize_no_stopwords(text):
    words = text.lower().split()
    return " ".join(w for w in words if w not in STOP_WORDS)


def normalize_stems(text):
    """Simplified stemming: remove common English suffixes."""
    suffixes = ["ing", "ed", "s", "es", "ly", "tion", "er", "est"]
    words = text.lower().split()
    result = []
    for word in words:
        for suffix in suffixes:
            if word.endswith(suffix) and len(word) > len(suffix) + 2:
                word = word[:-len(suffix)]
                break
        result.append(word)
    return " ".join(result)


strategies = {
    "basic": normalize_basic,
    "stopwords": normalize_no_stopwords,
    "stems": normalize_stems,
}

sample = episodes[11]["description"]
print(f"Original: {sample}\n")

for name, func in strategies.items():
    print(f"{name:<12}: {func(sample)}")
````

### 2.3 Passing Functions as Arguments

A function can receive another function as a parameter. This lets us
build flexible processing pipelines.

````python
def process_descriptions(episode_list, normalization_func):
    """Apply a normalization function to every episode description."""
    return [normalization_func(ep["description"]) for ep in episode_list]


sample_eps = episodes[:3]

print("=== Basic ===")
for text in process_descriptions(sample_eps, normalize_basic):
    print(f"  {text[:70]}...")

print("\n=== Without stop words ===")
for text in process_descriptions(sample_eps, normalize_no_stopwords):
    print(f"  {text[:70]}...")
````

### 3.2 Creating a Module

Any `.py` file is a module. The pattern `if __name__ == "__main__":` lets a
file work both as a runnable script **and** as an importable module.

We created a real module at `topic3-module-package/friends_utils.py`:

```python
# friends_utils.py

def format_code(season, episode):
    return f"S{season:02d}E{episode:02d}"

def episode_summary(ep):
    code = format_code(ep["season"], ep["episode"])
    return f"{code} - {ep['title']} [{ep['category']}]"

def filter_by_category(episodes, category):
    return [ep for ep in episodes if ep["category"] == category]

if __name__ == "__main__":
    print(format_code(1, 1))  # only runs when executed directly
```

- If you **run** `python friends_utils.py` → it prints `S01E01`
- If you **import** `from friends_utils import format_code` → it does **not** print anything

Let's import and use it:

````python
import sys
sys.path.insert(0, "topic3-module-package")

from friends_utils import format_code, episode_summary, filter_by_category

print("format_code(1, 7):", format_code(1, 7))
print("format_code(10, 17):", format_code(10, 17))

print("\nepisode_summary:")
print(episode_summary(episodes[0]))

print("\nfilter_by_category (Career):")
for ep in filter_by_category(episodes, "Career"):
    print(f"  {episode_summary(ep)}")
````

## 4. Standard Library Modules

Python includes many useful modules out of the box. Here are the most
relevant ones for data work.

### 4.1 `collections.Counter`

Counts elements automatically — perfect for word frequencies.

````python
from collections import Counter

category_counts = Counter(ep["category"] for ep in episodes)
print("Category distribution:")
for cat, count in category_counts.most_common():
    print(f"  {cat:<12} {count:>2}  {'#' * count}")

all_words = []
for ep in episodes:
    all_words.extend(ep["description"].lower().split())

print("\nTop 10 words across all descriptions:")
for word, count in Counter(all_words).most_common(10):
    print(f"  {word:<12} {count:>3}")
````

### 4.2 `collections.defaultdict`

Like a regular dictionary, but provides a default value for missing keys.

````python
from collections import defaultdict

episodes_by_category = defaultdict(list)

for ep in episodes:
    episodes_by_category[ep["category"]].append(ep["title"])

for cat in sorted(episodes_by_category):
    titles = episodes_by_category[cat]
    print(f"{cat} ({len(titles)}):")
    for t in titles:
        print(f"  - {t}")
    print()
````

### 4.3 `collections.namedtuple`

A lightweight alternative to a class — creates tuple-like objects
with named fields.

````python
from collections import namedtuple

EpisodeRecord = namedtuple("EpisodeRecord", ["season", "episode", "title", "category"])

records = [EpisodeRecord(ep["season"], ep["episode"], ep["title"], ep["category"])
           for ep in episodes[:5]]

for rec in records:
    print(f"S{rec.season:02d}E{rec.episode:02d} - {rec.title} [{rec.category}]")
````

### 4.4 `itertools`

Efficient looping utilities.

````python
from itertools import chain, combinations, groupby

romance_titles = [ep["title"] for ep in episodes if ep["category"] == "Romance"]
family_titles = [ep["title"] for ep in episodes if ep["category"] == "Family"]

print("chain() — merge two lists into one iterator:")
for title in chain(romance_titles[:2], family_titles[:2]):
    print(f"  {title}")

print(f"\ncombinations() — all pairs from first 4 titles:")
short_titles = [ep["title"] for ep in episodes[:4]]
for a, b in combinations(short_titles, 2):
    print(f"  {a}  +  {b}")

print(f"\ngroupby() — episodes grouped by season:")
sorted_eps = sorted(episodes, key=lambda ep: ep["season"])
for season, group in groupby(sorted_eps, key=lambda ep: ep["season"]):
    titles = [ep["title"] for ep in group]
    print(f"  Season {season:>2}: {len(titles)} episode(s)")
````

### 4.5 `datetime`

Working with dates and times.

````python
from datetime import datetime, timedelta

pilot_air_date = datetime.strptime("1994-09-22", "%Y-%m-%d")
finale_air_date = datetime.strptime("2004-05-06", "%Y-%m-%d")

print(f"Pilot aired : {pilot_air_date.strftime('%B %d, %Y')}")
print(f"Finale aired: {finale_air_date.strftime('%B %d, %Y')}")

run_duration = finale_air_date - pilot_air_date
print(f"Show ran for: {run_duration.days} days ({run_duration.days // 365} years)")

today = datetime.now()
since_finale = today - finale_air_date
print(f"\nDays since finale: {since_finale.days}")
print(f"That is approximately {since_finale.days // 365} years ago")
````
